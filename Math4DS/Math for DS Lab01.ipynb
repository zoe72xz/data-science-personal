```python
import matplotlib.pyplot as plt
import matplotlib.image as mimg

img0 = mimg.imread("/content/drive/MyDrive/Colab Notebooks/1.jpg")
img = img0
```

```python
from scipy import ndimage
rimg = ndimage.rotate(img, 90)
```

```python
plt.imshow(rimg)
plt.axis("off")
plt.show()
```

```python
#Start by discovering the dimensions or shape of the image.
image_height, image_width, num_channels = rimg.shape
print(f"Image Dimensions - Height: {image_height}, Width: {image_width}, Channels: {num_channels}")
```

```python
# To accomplish image manipulation, we'll need to import a package for array operations.
# We recommend using NumPy (Numerical Python) for its versatile capabilities.
import numpy as np
```

```python
# Create a new empty array of the same size as the previous image
new_image = np.empty_like(rimg)

# Modify one of the color channels of the new image using the intensity values from the previous image
new_image[:, :, 0] = rimg[:, :, 0] * 0.5  # Modify the Red channel (example: cut the intensity by half)
new_image[:, :, 1] = rimg[:, :, 1]  # Green channel remains the same
new_image[:, :, 2] = rimg[:, :, 2]  # Blue channel remains the same

# Create arrays with zeros for the other two color channels (Green and Blue)
new_image[:, :, 1] = 0  # Set Green channel to zero
new_image[:, :, 2] = 0  # Set Blue channel to zero

# Display the newly created image and remove axis
plt.imshow(new_image)
plt.axis("off")
plt.show()
```

```python
# Get the dimensions of the original image
height, width, _ = rimg.shape

# Create an empty grayscale image array with the same dimensions
grayscale_image = np.empty((height, width))

# Populate the grayscale image with pixel values (weighted average of color intensities)
for i in range(height):
    for j in range(width):
        # Calculate grayscale value using weighted average
        grayscale_value = 0.2989 * rimg[i, j, 0] + 0.5870 * rimg[i, j, 1] + 0.1140 * rimg[i, j, 2]
        grayscale_image[i, j] = grayscale_value

# Display the newly generated grayscale image
plt.imshow(grayscale_image, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()
```
